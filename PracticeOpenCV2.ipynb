```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show(img,size=5):
  aspect_ratio=img.shape[0]/img.shape[1]
  plt.figure(figsize=(size,aspect_ratio*size))
  plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
ben=np.zeros((1000,1000,3),np.uint8)
newimg=cv2.rectangle(ben,(200,200),(600,600),(200,100,0),10)
show(newimg)
nimg=cv2.rectangle(ben,(200,200),(600,600),(200,0,100),-10)
show(nimg)
```

```python
canvas=np.zeros((1000,1000,3),np.uint8)
k=cv2.circle(canvas,(500,500),200,(200,100,0),5)
show(k)
```

```python
img=np.zeros((1000,1000,3),np.uint8)
points=np.array([[100,850],[400,350],[700,500],[600,750]],np.int32)
points=points.reshape((-1,1,2))
cv2.polylines(img,[points],True,(0,0,255),6)
show(img)
```

```python
blackScreen=np.zeros((1000,1000,3),np.uint8)
cv2.putText(blackScreen,"Good Morning Everyone",(200,200),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,2,(20,100,200),3)
show(blackScreen)
```

Translations and Rotations

affine transform - simply pushes image left or right

```python
panda=cv2.imread("/content/cutePanda.jpg")
T=[[1,0,panda.shape[1]/4],[0,1,panda.shape[0]/4]]
T=np.array(T,dtype=np.float32)
newPanda=cv2.warpAffine(panda,T,(panda.shape[1],panda.shape[0]))
show(newPanda)
```

```python
T=[[1,0,panda.shape[1]/4],[0.3,1,panda.shape[0]/4]]
T=np.array(T,dtype=np.float32)
newPanda=cv2.warpAffine(panda,T,(panda.shape[1],panda.shape[0]))
show(newPanda)
```

```python
T=[[0.5,0.3,panda.shape[1]/4],[0,1,panda.shape[0]/4]]
T=np.array(T,dtype=np.float32)
newPanda=cv2.warpAffine(panda,T,(panda.shape[1],panda.shape[0]))
show(newPanda)
```

```python
R=cv2.getRotationMatrix2D((panda.shape[1]/2,panda.shape[0]/2),45,1)
newPanda=cv2.warpAffine(panda,R,(panda.shape[1],panda.shape[0]))
show(newPanda)
```

```python
rotated_image=cv2.transpose(panda)
i=cv2.transpose(rotated_image)
show(i)
```

```python
scaledPanda=cv2.resize(panda,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)
show(scaledPanda)
```

```python
startcol,startrow=.25*panda.shape[1],.25*panda.shape[0]
endcol,endrow=.75*panda.shape[1],.75*panda.shape[0]
cropped=panda[int(startrow):int(endrow),int(startcol):int(endcol)]
show(cropped)
```
